**Quicksort with non-random pivot**


```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quicksort(less) + [pivot] + quicksort(greater)
```

**Quicksort with random pivot**


```python
import random

def quicksort_random_pivot(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = random.choice(arr)
        less = [x for x in arr if x < pivot]
        equal = [x for x in arr if x == pivot]
        greater = [x for x in arr if x > pivot]
        return quicksort_random_pivot(less) + equal + quicksort_random_pivot(greater)
```

**2a)Best Case**

```python
def generate_best_case_input(n):
    return list(range(1, n + 1))

best_case_inputs = [generate_best_case_input(n) for n in range(1, 101)]
```

**2b) Worst Case**

```python
def generate_worst_case_input(n):
    return list(range(n, 0, -1))

worst_case_inputs = [generate_worst_case_input(n) for n in range(1, 101)]
```

**2 c) Average Case:**

```python
import random

def generate_average_case_input(n):
    return [random.randint(1, 1000) for _ in range(n)]

average_case_inputs = [generate_average_case_input(n) for n in range(1, 101)]
```

```python
import timeit
import matplotlib.pyplot as plt

def benchmark_quicksort_non_random_pivot(inputs, title):
    times = []
    for arr in inputs:
        time = timeit.timeit(lambda: quicksort(arr.copy()), number=10)
        times.append(time)
    
    plt.plot(range(1, len(inputs) + 1), times, label=title)

benchmark_quicksort_non_random_pivot(best_case_inputs, 'Best Case')
benchmark_quicksort_non_random_pivot(worst_case_inputs, 'Worst Case')
benchmark_quicksort_non_random_pivot(average_case_inputs, 'Average Case')

plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.legend()
plt.title('Quicksort Non-Random Pivot Benchmarks')
plt.show()
```
